```python
from sklearn import datasets
import pandas as pd
import numpy as np

```

```python
def step_gradient(points,learning_rate,m,c):
    m_slope =0
    c_slope = 0
    M=len(points)
    for i in range(M):
        x=points[i,0]
        y=points[i,1]
        m_slope += (-2/M)*(y - m * x-c) * x
        c_slope += (-2/M) * (y - m * x - c)
    new_m = m - learning_rate* m_slope
    new_c = m - learning_rate* c_slope
    return new_m ,new_c
    
        
```

```python
def gd(points,learning_rate,num_iteration):
    m=0
    c=0
    for i in range(num_iteration):
        m,c = step_gradient(points,learning_rate,m,c)
        print(i,"cost:",cost(points,m,c))
    return m,c   
```

```python
def cost(points,m,c):
    total_cost = 0
    M=len(points)
    for i in range(M):
        x=points[i,0]
        y=points[i,1]
        total_cost +=  (1/M) * ((y - m * x - c)**2)
    return total_cost   
```

```python
def run():
    data = np.loadtxt("0000000000002417_training_boston_x_y_train.csv",delimiter=',')
    learning_rate = 0.000001
    num_iteration = 1000
    m,c = gd(data,learning_rate,num_iteration)
    print(m,c)
```

```python
run()
```
